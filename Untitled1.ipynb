```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules

dataset = [
    ['A','B','C','D'],
    ['A','C','E'],
    ['B','C','D','E'],
    ['A','B','D','E'],
    ['A','B','C']
]

df = pd.DataFrame(dataset)

df
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)
df_hot = pd.DataFrame(te_data, columns=te.columns_)
df_hot
```

```python
frequent_apriori = apriori(df_hot, min_support=0.5, use_colnames=True)
frequent_apriori
```

```python
frequent_fpgrowth = fpgrowth(df_hot, min_support=0.5, use_colnames=True)
frequent_fpgrowth
```

```python
rules = association_rules(frequent_apriori, metric="confidence", min_threshold=0.7)
rules
```

```python
data = {
    "Department": ["Sales","Sales","Sales","IT","IT","IT","IT","HR","HR","Admin","Admin"],
    "Status": ["Senior","Junior","Junior","Junior","Senior","Junior","Senior","Senior","Junior","Senior","Junior"],
    "Age": ["31-35","26-30","31-35","21-25","36-40","26-30","41-45","36-40","31-35","46-50","26-30"],
    "Salary": ["51-55","26-30","31-35","41-45","61-65","46-50","61-65","46-50","36-40","36-40","26-30"],
    "Count": [25,35,45,15,8,5,7,12,6,3,7]
}

df = pd.DataFrame(data)
df
```

```python
from math import log2

def entropy(p1, p2):
    total = p1+p2
    if p1==0 or p2==0:
        return 0
    return -(p1/total)*log2(p1/total) - (p2/total)*log2(p2/total)

entropy(25+8+12+3+7, 35+45+15+5+6+7)
```

```python
# Count totals
senior_count = df[df["Status"]=="Senior"]["Count"].sum()
junior_count = df[df["Status"]=="Junior"]["Count"].sum()
total = senior_count + junior_count

P_senior = senior_count/total
P_junior = junior_count/total

P_senior, P_junior
```

```python
# Likelihoods for (IT, 26-30, 46-50)
import numpy as np

def cond_prob(attribute, value, status):
    sub = df[(df[attribute]==value)&(df["Status"]==status)]
    return sub["Count"].sum() / df[df["Status"]==status]["Count"].sum()

IT_26_30_senior = cond_prob("Department","IT","Senior") * cond_prob("Age","26-30","Senior") * cond_prob("Salary","46-50","Senior")
IT_26_30_junior = cond_prob("Department","IT","Junior") * cond_prob("Age","26-30","Junior") * cond_prob("Salary","46-50","Junior")

Posterior_senior = IT_26_30_senior * P_senior
Posterior_junior = IT_26_30_junior * P_junior

Posterior_senior, Posterior_junior
```
